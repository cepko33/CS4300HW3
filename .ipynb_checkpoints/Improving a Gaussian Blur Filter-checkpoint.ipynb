## Thesis
Through the combination of Gaussian filtering and matrix convolution, a repeatable, configurable, and comprehensive implementation of the Gaussian Blur can be constructed.

## Introduction
The Gaussian Blur is a widely utilized blurring algorithm that is implemented by applying a low-pass Gaussian filter to every dimension of a given matrix. In the field of image processing, the Gaussian Blur is a staple, and it provides a pixel normalization that results in a smooth and organic blurring effect.

A graphed Gaussian Function provides a bell-like curve and is expressed in the form: 
$$ f(x) = a * exp (- \frac{(x-b)^2}{2c^2}) $$

Where $a$ is the height of the curve, $b$ is the x offset, and $c$ is the width of the curve.[0]



````python
%matplotlib inline
import math
import matplotlib.pyplot as plt
import numpy
from scipy import ndimage as ndi
from skimage import color
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
import matplotlib.image as mpimg
from skimage import img_as_float, img_as_ubyte, img_as_uint, img_as_bool
import glob # for bulk file import

plt.rcParams['figure.figsize'] = 25, 10

# Standard Gaussian Function

def gauss_one(x, a, b, c):
    return a * math.exp(-1 * float((x - b)**2)/float(2*c**2))

def gauss_one_ex(x):
    return gauss_one(x, 1, 0, 1)

# Modified (improved) Gaussian Function

def gauss_one_imp(x, a, b, c):
    return a * math.exp(-1 * float((x - b)**2)/float(2*c**2) - math.sqrt((x**2)/4))

def gauss_one_imp_ex(x):
    return gauss_one_imp(x, 1, 0, 1)

x = numpy.linspace(-2,2,100)

gaussonevector = numpy.vectorize(gauss_one_ex)
gaussonevector_imp = numpy.vectorize(gauss_one_imp_ex)

plt.title("Signal of a One Dimensional Gaussian", fontsize=30)
plt.plot(x, gaussonevector(x))
plt.show()

plt.title("Traditional Gaussian Function vs. Modified Gaussian Function", fontsize=30)
plt.plot(x, gaussonevector(x))
plt.plot(x, gaussonevector_imp(x))
plt.show()
````

This generalized function can be cast into 3 dimensions by rotating the curve around its peak in the Z dimension, and can also be expressed in the form:

$$ f(x,y) = A * exp ( - (\frac{(x - x_0)^2}{2\sigma^2_x} + \frac{(y - y_0)^2}{2\sigma^2_y}) $$

Where A is the amplitude/height of the curve, $x_0,y_0$ is the center point on the Horizontal plane, and $\sigma_x,\sigma_y$ describe the x,y spread of the curve[1].


````python

# Standard Two-Dimensional Gaussian Function

def gauss_two(x,y,sigx,sigy, a):
    return a * math.exp(-1 * (((x**2)/2*sigx**2) + (y**2)/2*sigy**2))

def gauss_two_ex(x,y):
    return gauss_two(x,y,1,1,1)

# Modified (improved) Two-Dimensional Gaussian Function

def gauss_two_imp(x,y,sigx,sigy, a):
    return a * math.exp(-1 * (((x**2)/2*sigx**2) + (y**2)/2*sigy**2) - math.sqrt((x/4)**2 + (y/4)**2))

def gauss_two_imp_ex(x,y):
    return gauss_two_imp(x,y,1,1,1)

X = numpy.arange(-1.5, 1.5, 0.01)
Y = numpy.arange(-1.5, 1.5, 0.01)
X, Y = numpy.meshgrid(X, Y)

colors = cm.rainbow

fig = plt.figure()
ax = fig.gca(projection='3d')

gausstwovector = numpy.vectorize(gauss_two_ex)

ax.plot_surface(X,Y, gausstwovector(X,Y),cmap=colors)
plt.title('Surface Plot of a Standard Two Dimensional Gaussian', fontsize=30)
plt.show()

fig = plt.figure()
ax = fig.gca(projection='3d')

gausstwovectorimp = numpy.vectorize(gauss_two_imp_ex)

ax.plot_surface(X,Y, gausstwovectorimp(X,Y),cmap=colors)
plt.title('Surface Plot of Modified Two Dimensional Gaussian', fontsize=30)
plt.show()
````

## Method
The two-dimensional Gaussian Function can provide a way to average the pixels in a specified area by using the Z value as the weighting amount. Using a convolve operation which casts a kernel, in this case a matrix that represents a two-dimensional Gaussian function, onto the image, each pixel of the resulting image will be a weighted averaging of the pixels around it.

The basic algorithm to compute the Gaussian Blur is
1. Generate a two-dimensional convolution matrix based on a Gaussian Function
2. Convolve the matrix as a kernel over the given image

As an improvement, the Gaussian Function is tweaked by intersecting the normal of the position with a Square Root function, resulting in a coordinate weighting that factors adjacent pixels more favorably than those farther out. 

The modified Gaussian Functions are calculated using the following functions:

$$ f(x) = a * exp (- \frac{(x-b)^2}{2c^2} - \sqrt{\frac{x}{4}^2}) $$

$$ f(x,y) = A * exp ( - (\frac{(x - x_0)^2}{2\sigma^2_x} + \frac{(y - y_0)^2}{2\sigma^2_y} - \sqrt{\frac{x}{4}^2 + \frac{y}{4}^2}) $$



````python
vecx = numpy.arange(-1, 1, .1)
vecy = numpy.arange(-1, 1, .1)

matrix = []
for x in vecx:
    row = []
    for y in vecy:
        row.append(gauss_two_ex(x,y))
    matrix.append(row)

#print matrix
#gausstwovector(vecx,vecy)

plt.figure()
plt.subplot(1, 2, 1)
plt.title("Gaussian", fontsize=30)

plt.imshow(matrix, cmap=cm.gray)


matrix = []
for x in vecx:
    row = []
    for y in vecy:
        row.append(gauss_two_imp_ex(x, y))
    matrix.append(row)
    
plt.subplot(1, 2, 2)
plt.title("Modified Gaussian", fontsize=30)
plt.imshow(matrix, cmap=cm.gray)
````

This application of a black and white gradient to the original and modified Gaussian Function implementations reveals that the modified function appears to have a more spherical appearance, and gives a more concentrated weight to pixels immediately surrounding the apex.

## Code

````python
plt.rcParams['figure.figsize'] = 25, 10

imgpaths = glob.glob("./images/*.jpg") + glob.glob("./images/*.png") + glob.glob("./images/*.JPG")
imgset = [img_as_float(mpimg.imread(x)) for x in imgpaths]

# Display thumbnails of the images to ensure loading
plt.figure()
for i,img in enumerate(imgset):
    plt.subplot(len(imgset), 1, i+1)
    plt.imshow(color.rgb2grey(img), cmap = 'gray')
````

````python
def generate_kernel(steps, size_x, size_y):
    vec = numpy.arange(-1.0,1.0,1.0/(steps/2))
    matrix = []
    for x in vec:
        row = []
        for y in vec:
            row.append(gauss_two_imp(x,y,size_x,size_y,1))
        matrix.append(row)
    return matrix

plt.figure()
plt.suptitle("Modified Gaussian Image Blur Test", fontsize=30)
for i, img in enumerate(imgset):
    imgbw = color.rgb2grey(img)
    plt.subplot(len(imgset), 1, i+1)
    plt.imshow(ndi.convolve(imgbw,generate_kernel(20,1,1)), cmap = 'gray')
````

````python
plt.figure()
plt.suptitle("Modified Gaussian Image Blur With Vertical Bias", fontsize=30)
for i, img in enumerate(imgset):
    imgbw = color.rgb2grey(img)
    plt.subplot(len(imgset), 1, i+1)
    plt.imshow(ndi.convolve(imgbw,generate_kernel(30,0.2,10)), cmap = 'gray')
````

By creating a pipeline from Gaussian Matrix to convolution, a reliable Gaussian Blur function can be constructed.

## Results


````python
unmodified = color.rgb2grey(imgset[1])
with_gaussian = ndi.convolve(color.rgb2grey(imgset[1]), generate_kernel(20,1,1))

plt.figure()
plt.subplot(1, 2, 1)
plt.title("Histogram of an Unmodified Image", fontsize=20)
for col, channel in zip('rgb',numpy.rollaxis(unmodified,axis=-1)):
    plt.hist(channel.flatten(), bins=50, color=col,alpha=0.3)

plt.subplot(1, 2, 2)
plt.title("Histogram of Image Processed with Modified Gaussian", fontsize=20)
for col, channel in zip('rgb',numpy.rollaxis(with_gaussian, axis=-1)):
    plt.hist(channel.flatten(), bins=50, color=col,alpha=0.3)

plt.show()

````

The histograms above best display the effect of the Gaussian Blur on one of the greyscale images. The left histogram shows the frequencies before the blurring, and the right histogram shows the frequencies after the blurring.

In the rightG histogram, all of the peaks and bumps in frequencies in every channel have been softened. Noise in the RGB channels has been reduced significantly, providing a unified greyscale signal.

````python
plt.figure()

hist_unmodified = numpy.histogram(img_as_ubyte(unmodified), bins=numpy.arange(0, 256))
hist_gaussian = numpy.histogram(with_gaussian, bins=numpy.arange(0, 256))

plt.title("Comparison of Histogram Frequencies", fontsize=20)
plt.plot(hist_unmodified[1][:-1], hist_unmodified[0], lw=2, label="Unmodified Image")
plt.plot(hist_gaussian[1][:-1], hist_gaussian[0], lw=2, label="Image With Gaussian Blur")
plt.legend(fontsize=20)
plt.show()
````

Better illustrated in this graph is the fact that the grayscale balance in the Gaussian filter concentrates the frequencies towards the middle

````python
import skimage.filter as filters

for i, img in enumerate(imgset):
    plt.figure()
    imgbw = color.rgb2grey(img)
    plt.subplot(1, 2, 1)
    plt.title("Sci-Kit Gaussian", fontsize=20)
    plt.imshow(filters.gaussian_filter(imgbw, 5,multichannel=False), cmap = 'gray')
    plt.subplot(1, 2, 2)
    plt.title("Modified Gaussian", fontsize=20)
    plt.imshow(ndi.convolve(imgbw,generate_kernel(20,1,1)), cmap = 'gray')
````

## Discussion

The intersection of the Gaussian Function and Square Root provides a helpful new level of control that can be applied in general Gaussian Blur filters. While it is difficult to say whether the modified Gaussian Filter improved upon the original algorithm, it definitely has a distinct difference in the resulting image. The values of immediately adjacent pixels are much more heavily weighted than those just a few pixel-lengths away. Having the ability to modify the rate and type of weight change in a Gaussian Blur provides the user with more interesting choices in the final effect.
